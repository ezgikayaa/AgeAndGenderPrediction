```python
from sklearn.metrics import mean_squared_error
def sigmoid(s):u

    return 1/(1 + np.exp(-s))

def sigmoid_derv(s):
    return s * (1 - s)

def softmax(X):
    exps = np.exp(X)
    return exps / np.sum(exps)

def relu(x):
    return np.maximum(0,x)

def reluDerivative(x):
    x[x<=0] = 0
    x[x>0] = 1
    return x

def LReLU_function( signal, leakage = 0.01 ):
    
    output = np.copy( signal )
    output[ output < 0 ] *= leakage
        
    return output

def LReLU_der( signal, leakage = 0.01 ):
    return np.clip(signal > 0, leakage, 1.0)

def tanh(x):
    return np.tanh(x)

def tanh_deriv(x):
    return 1.0 - np.tanh(x)**2

def mean_error(pred, real):
    n_samples = real.shape[0]
    res = pred - real
    return res/n_samples

def error(pred, real):
    n_samples = real.shape[0]
    logp = - np.log(pred[np.arange(n_samples), real.argmax(axis=1)])
    loss = np.sum(logp)/n_samples
    return loss


def get_acc(x, y,validation):
    TP = 0
    TN = 0
    FN = 0
    FP = 0
    
    for xx,yy in zip(x, y):
        s = model.predict(xx)
        ss = np.argmax(yy)
        if validation==1:
            print(s)
        if s == 1 and ss == 1:
            TN +=1
        if s == 0 and ss == 0:
            TP +=1
        if s == 1 and ss == 0:
            FN+=1
        if s == 0 and ss == 1:
            FP+=1
            
    return (TP+TN)/len(x)*100, TP/(TP+FP) , TP/(TP+FN) 

class NN:
    def __init__(self, x, y):
        self.x = x
        neurons =500
        self.lr = 0.05
        ip_dim = x.shape[1]
        op_dim = y.shape[1]

        self.w1 = np.random.randn(ip_dim, neurons)
        self.b1 = np.zeros((1, neurons))
        self.w2 = np.random.randn(neurons, neurons)
        self.b2 = np.zeros((1, neurons))
        self.w3 = np.random.randn(neurons, op_dim)
        self.b3 = np.zeros((1, op_dim))
        self.y = y

    def feedforward(self):
        z1 = np.dot(self.x, self.w1) + self.b1
        self.a1 = tanh(z1)
        z2 = np.dot(self.a1, self.w2) + self.b2
        self.a2 = tanh(z2)
        z3 = np.dot(self.a2, self.w3) + self.b3
        self.a3 = sigmoid(z3)

    def backprop(self):
        loss = error(self.a3, self.y)
        print('Error :', loss)
        a3_delta = mean_error(self.a3, self.y) # w3
        z2_delta = np.dot(a3_delta, self.w3.T)
        a2_delta = z2_delta * tanh_deriv(self.a2) # w2
        z1_delta = np.dot(a2_delta, self.w2.T)
        a1_delta = z1_delta * tanh_deriv(self.a1) # w1

        self.w3 -= self.lr * np.dot(self.a2.T, a3_delta)
        self.b3 -= self.lr * np.sum(a3_delta, axis=0, keepdims=True)
        self.w2 -= self.lr * np.dot(self.a1.T, a2_delta)
        self.b2 -= self.lr * np.sum(a2_delta, axis=0)
        self.w1 -= self.lr * np.dot(self.x.T, a1_delta)
        self.b1 -= self.lr * np.sum(a1_delta, axis=0)


    def predict(self, data):
        self.x = data
        self.feedforward()
        return self.a3.argmax()
    
model = NN(x_train/255.0, np.array(y_train))

epochs =2000

for x in range(epochs):
    model.feedforward()
    model.backprop()
    print("Epoch  : "+str(x)+"/" +str(epochs) )
    #255 e böl
    

acc, precision, recall = get_acc(x_train/255, np.array(y_train),0)
print("Training accuracy : ",acc )
print("Training precision : ",precision )
print("Training recall : ",recall )

acc, precision, recall = get_acc(x_val/255, np.array(y_val),0)
print("Test accuracy : ",acc )
print("Test precision : ",precision )
print("Test recall : ",recall )


```
